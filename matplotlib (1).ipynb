### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python data visualization library built on top of Matplotlib, providing a high-level interface for drawing attractive and informative statistical graphics.

 Here are five common plots that can be created using Seaborn, along with their uses:

### 1. **Scatter Plot (`sns.scatterplot`)**
   - **Use:** A scatter plot is used to visualize the relationship between two continuous variables. Each point in the plot represents an observation from the dataset. It's helpful in identifying correlations, trends, and potential outliers.

### 2. **Bar Plot (`sns.barplot`)**
   - **Use:** A bar plot is used to display the distribution of a categorical variable and the comparison of means (or other summary statistics) between different categories. The height of each bar represents the value of the variable for each category. It's useful for comparing the central tendency of different groups.

### 3. **Box Plot (`sns.boxplot`)**
   - **Use:** A box plot (or whisker plot) is used to show the distribution of a dataset by displaying the data's minimum, first quartile, median, third quartile, and maximum. It’s particularly useful for identifying outliers and understanding the spread and skewness of the data.

### 4. **Heatmap (`sns.heatmap`)**
   - **Use:** A heatmap is used to visualize matrix-like data, where each cell in the matrix is color-coded based on its value. It's commonly used to display the correlation matrix of variables, which helps in identifying relationships and patterns within the data.

### 5. **Histogram (`sns.histplot`)**
   - **Use:** A histogram is used to display the distribution of a single continuous variable by dividing the data into bins and plotting the frequency of observations in each bin. It’s useful for understanding the underlying distribution, central tendency, and spread of the data.

These plots can be customized extensively in Seaborn to enhance data analysis and presentation.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Box plot for 'pclass' vs 'age'
sns.boxplot(data=titanic, x='pclass', y='age', ax=axes[0])
axes[0].set_title('Box Plot of Age by Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')

# Box plot for 'pclass' vs 'fare'
sns.boxplot(data=titanic, x='pclass', y='fare', ax=axes[1])
axes[1].set_title('Box Plot of Fare by Passenger Class')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
c = sns.load_dataset('diamonds')
sns.histplot(x=c.price,hue='cut',data=c,multiple='stack')
```
